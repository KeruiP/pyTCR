Step 3 - Create the standardized format (column content and column names)
1.   Sample name (sample)
2.   Clone frequency (freq)
3.   Clone reads count (#count)
4.   CDR3 amino acid clonotype (cdr3aa)
5.   CDR3 nucleotide (cdr3nt)
6.   V gene (v)
7.   D gene (d)
8.   J gene (j)
9.   Optional - Sample feature (here we use hospitalization as the clinical feature)

```
df.drop('sample_name', axis=1, inplace=True)
df.columns = ['sample', 'hospitalization' , 'freq' , '#count' , 'cdr3aa' , 'cdr3nt' , 'v' , 'd' , 'j']
df
```

Prepare two dataframes for merging and analyzing the overlapping

```
df_samples = df.copy()

#if there are more than one v,j, leave only the first one
df_samples['v'] = df_samples['v'].str.replace("(,).*","")
df_samples['j'] = df_samples['j'].str.replace("(,).*","")

df_compare = pd.merge(df_samples, df_samples, on=['cdr3nt','cdr3aa','v','d','j'], suffixes = ['_1', '_2'])

df_compare['#count_1**2'] = df_compare['#count_1'] * df_compare['#count_1']
df_compare['#count_2**2'] = df_compare['#count_2'] * df_compare['#count_2']
df_compare['#count_1*2'] = df_compare['#count_1'] * df_compare['#count_2']
```

Overlap analysis 1.1 - Jaccard index (result table)

```
samples = list(set(df['sample']))

N = len(samples)

#create an empty dataframe for storing results
df_Jaccard_index = pd.DataFrame(columns = ['sample1','sample2','Jaccard_index'])

for i in range(0,N):
  for j in range(i+1,N):

    sample1 = samples[i]
    sample2 = samples[j]

    df_sample1 = df_samples.loc[(df_samples['sample'] == sample1)]
    df_sample2 = df_samples.loc[(df_samples['sample'] == sample2)]

    df_overlap = df_compare.loc[(df_compare['sample_1'] == sample1) & (df_compare['sample_2'] == sample2)]
    Jaccard_index = df_overlap.shape[0]/(df_sample1.shape[0]+df_sample2.shape[0]-df_overlap.shape[0])

    #store the results 
    df_temp1 = pd.DataFrame(data={'sample1':sample1, 'sample2':sample2, 'Jaccard_index':Jaccard_index}, columns=['sample1', 'sample2', 'Jaccard_index'],index=[0])
    df_temp2 = pd.DataFrame(data={'sample1':sample2, 'sample2':sample1, 'Jaccard_index':Jaccard_index}, columns=['sample1', 'sample2', 'Jaccard_index'],index=[0])
    df_temp = pd.concat([df_temp1,df_temp2])

    df_Jaccard_index = pd.concat([df_Jaccard_index,df_temp])

df_Jaccard_index 
```

Overlap analysis 1.2 - Jaccard index (heatmap)

```
#set aesthetics
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(40,25))
sns.set_style("white")
sns.set_context("talk")

#generate axes
df_Jaccard_index['Jaccard_index'] = df_Jaccard_index['Jaccard_index'].astype(float)
result = df_Jaccard_index.pivot(index='sample2',columns='sample1',values='Jaccard_index')
ax = sns.heatmap(result, cmap='coolwarm')
ax.set_xlabel('sample1',fontsize=25)
ax.set_ylabel('sample2',fontsize=25)
sns.despine()
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.title('Jaccard_index', fontsize=30)
```

```
samples = list(set(df['sample']))

N = len(samples)

#create an empty dataframe for storing results
df_overlap_coefficient = pd.DataFrame(columns = ['sample1','sample2','overlap_coefficient'])

for i in range(0,N):
  for j in range(i+1,N):

    sample1 = samples[i]
    sample2 = samples[j]

    df_sample1 = df_samples.loc[(df_samples['sample'] == sample1)]
    df_sample2 = df_samples.loc[(df_samples['sample'] == sample2)]

    df_overlap = df_compare.loc[(df_compare['sample_1'] == sample1) & (df_compare['sample_2'] == sample2)]

    overlap_coefficient = df_overlap.shape[0]/min(df_sample1.shape[0], df_sample2.shape[0])

    #store the results 
    df_temp1 = pd.DataFrame(data={'sample1':sample1, 'sample2':sample2, 'overlap_coefficient':overlap_coefficient}, columns=['sample1', 'sample2', 'overlap_coefficient'],index=[0])
    df_temp2 = pd.DataFrame(data={'sample1':sample2, 'sample2':sample1, 'overlap_coefficient':overlap_coefficient}, columns=['sample1', 'sample2', 'overlap_coefficient'],index=[0])
    df_temp = pd.concat([df_temp1,df_temp2])

    df_overlap_coefficient = pd.concat([df_overlap_coefficient,df_temp])

df_overlap_coefficient
```

Overlap analysis 3.1 - Morisita-Horn index (result table)

```
samples = list(set(df['sample']))

N = len(samples)

#create an empty dataframe for storing results
df_Morisita_Horm_index = pd.DataFrame(columns = ['sample1','sample2','Morisita_Horm_index'])

for i in range(0,N):
  for j in range(i+1,N):

    sample1 = samples[i]
    sample2 = samples[j]
    
    sum_sample_1_count = df_samples.loc[(df_samples['sample'] == sample1),'#count'].sum()
    sum_sample_2_count = df_samples.loc[(df_samples['sample'] == sample2),'#count'].sum()

    sum_sample_1	= df_compare.loc[(df_compare['sample_1'] == sample1) & (df_compare['sample_2'] == sample2),'#count_1**2'].sum()
    sum_sample_2	= df_compare.loc[(df_compare['sample_1'] == sample1) & (df_compare['sample_2'] == sample2),'#count_2**2'].sum()

    sum_count_product = df_compare.loc[(df_compare['sample_1'] == sample1) & (df_compare['sample_2'] == sample2),'#count_1*2'].sum()
 
    #calculate Morisita-Horn index
    step1 = (sum_sample_1/((sum_sample_1_count)**2)) + (sum_sample_2/((sum_sample_2_count)**2))
    step2 = step1 * sum_sample_1_count * sum_sample_2_count
    step3 = 2*sum_count_product
    Morisita_Horm_index = step3/step2

    #store the results 
    df_temp1 = pd.DataFrame(data={'sample1':sample1, 'sample2':sample2, 'Morisita_Horm_index':Morisita_Horm_index}, columns=['sample1','sample2','Morisita_Horm_index'],index=[0])
    df_temp2 = pd.DataFrame(data={'sample1':sample2, 'sample2':sample1, 'Morisita_Horm_index':Morisita_Horm_index}, columns=['sample1','sample2','Morisita_Horm_index'],index=[0])
    df_temp = pd.concat([df_temp1,df_temp2])

    df_Morisita_Horm_index = pd.concat([df_Morisita_Horm_index,df_temp])

df_Morisita_Horm_index
```

Overlap analysis 3.2 - Morisita-Horn index (heatmap)

```
#set aesthetics
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(40,25))
sns.set_style("white")
sns.set_context("talk")

#generate axes
df_Morisita_Horm_index['Morisita_Horm_index'] = df_Morisita_Horm_index['Morisita_Horm_index'].astype(float)
result = df_Morisita_Horm_index.pivot(index='sample2',columns='sample1',values='Morisita_Horm_index')
ax = sns.heatmap(result, cmap='coolwarm')
ax.set_xlabel('sample1',fontsize=20)
ax.set_ylabel('sample2',fontsize=20)
sns.despine()
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

plt.title('Morisita_Horm_index')
```

Overlap analysis 10.1 - Jensen-Shannon divergence of variable segment usage distributions (result table)

```
samples = list(set(df['sample']))

N = len(samples)

#create an empty dataframe for storing results
df_Jensen_Shannon_divergence = pd.DataFrame(columns = ['sample1','sample2','Jensen_Shannon_divergence'])

def kl_divergence(p, q): 
    return -np.sum(p*np.log2(q/p))

def js_divergence(p, q):
    m = (1/2)*(p + q)
    return (1/2)*kl_divergence(p, m) + (1/2)*kl_divergence(q, m)

for i in range(0,N):
  for j in range(i+1,N):

    sample1 = samples[i]
    sample2 = samples[j]

    df_sample1 = df_samples.loc[(df_samples['sample'] == sample1)]
    df_sample2 = df_samples.loc[(df_samples['sample'] == sample2)]

    df_sample1_v = df_sample1.groupby(['v'], as_index=False)['freq'].agg({'sumfreq_1':'sum'})
    df_sample2_v = df_sample2.groupby(['v'], as_index=False)['freq'].agg({'sumfreq_2':'sum'})

    df_JSD_combine = pd.merge(df_sample1_v, df_sample2_v, on=['v'])

    JSD_sample1 = df_JSD_combine[['sumfreq_1']].to_numpy()
    JSD_sample2 = df_JSD_combine[['sumfreq_2']].to_numpy()

    Jensen_Shannon_divergence = js_divergence(JSD_sample1, JSD_sample2)

    #store the results 
    df_temp1 = pd.DataFrame(data={'sample1':sample1, 'sample2':sample2, 'Jensen_Shannon_divergence':Jensen_Shannon_divergence}, columns=['sample1', 'sample2', 'Jensen_Shannon_divergence'],index=[0])
    df_temp2 = pd.DataFrame(data={'sample1':sample2, 'sample2':sample1, 'Jensen_Shannon_divergence':Jensen_Shannon_divergence}, columns=['sample1', 'sample2', 'Jensen_Shannon_divergence'],index=[0])
    df_temp = pd.concat([df_temp1,df_temp2])

    df_Jensen_Shannon_divergence = pd.concat([df_Jensen_Shannon_divergence,df_temp])

df_Jensen_Shannon_divergence
```
